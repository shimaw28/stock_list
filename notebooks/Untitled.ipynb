```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
year = 2018
url = f"https://www.ipokiso.com/company/{year}.html"
```

```python
res = requests.get(url)
```

```python
soup = BeautifulSoup(res.text)
```

```python
tables = pd.read_html(url)
```

```python
df = pd.DataFrame()
for i in range(0, len(tables), 2):
    table = pd.concat([tables[i], tables[i+1]], axis=1)
    df = pd.concat([df, table])
```

```python
df.head()
```

```python
df = df[~(df["企業名"] == "企業名")]
```

```python
df["企業名"] = df["企業名"].str.extract(r"\((.+)\)")
```

```python
df["上場日"] = f"{year}/" + df["上場日"]
```

```python
df.shape
```

```python
rates = []
for img in soup.select("img[alt]"):
    rate = img["alt"]
    if len(rate) == 1:
        rates.append(rate)
```

```python
len(rates)
```

```python
df["rate"] = rates
```
